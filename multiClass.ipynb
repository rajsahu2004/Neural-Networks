```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense
```

```python
classes = 4
m =100
centers = [[-5,2],[-2,-2],[1,2],[5,-2]]
std = 1
xTrain,yTrain = make_blobs(n_samples=m,centers=centers,cluster_std=std,random_state=30)
```

```python
print(f'Unique classes : {np.unique(yTrain)}')
print(f'Class representation : {yTrain[:10]}')
print(f'shape of xTrain : {xTrain.shape}, shape of yTrain : {yTrain.shape}')
```

```python
tf.random.set_seed(1234)
model = Sequential(
    [
        Dense(2,activation='relu',name='L1'),
        Dense(4,activation='linear',name='L2')
    ]
)
```

```python
model.compile(
    loss= tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    optimizer=tf.keras.optimizers.Adam(0.01)
)
model.fit(xTrain,yTrain,epochs=200)
```
